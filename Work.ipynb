```python
# loading data 
data = pd.read_csv('/Users/mac/Documents/Projects/P1/uploads/')
```

```python
data['DATE_OF_TRAN'] = pd.to_datetime(data['DATE_OF_TRAN'])
```

```python
categories = data['RPTCODE'].unique()
```

```python
# create different dataframe
data_A = data[data['RPTCODE']==categories[0]]
data_B = data[data['RPTCODE']==categories[1]]
data_I = data[data['RPTCODE']==categories[2]]
data_R = data[data['RPTCODE']==categories[3]]
```

```python
cols = data.columns
cols
```

```python
# Made Dictionary to store processed DataFrames
final_dfs = {}

# Process each category dynamically
for category in categories:
    filtered_data = data[data['RPTCODE'] == category]  # Filtering data by category
    total_amounts = filtered_data.groupby('DATE_OF_TRAN', as_index=False)['TRAN_AMT'].sum()  # Sum transactions per date

    # Add missing columns to match structure
    for col in ['SR', 'TRANID', 'ACCOUNT NUMBER', 'DR/CR', 'RPTCODE', 'ACCOUNT NAME']:
        total_amounts[col] = None  # Keep other columns empty in total row

    # Mark total rows
    total_amounts['Total'] = 'Total'
    filtered_data['Total'] = ''

    # Concatenate original data with total rows and maintain order
    df_final = pd.concat([filtered_data, total_amounts], ignore_index=False).sort_values(by=['DATE_OF_TRAN', 'Total'])

    # Drop helper column
    df_final.drop(columns=['Total'], inplace=True)

    # Reset index
    df_final.reset_index(drop=True, inplace=True)

    # Store the result in a dictionary
    final_dfs[category] = df_final

    
    
# merging data of all the categories after processing
merged_df = pd.concat(list(final_dfs.values()), ignore_index=True)
# Save to CSV
merged_df.to_csv(f"/Users/mac/Documents/Projects/P1/processed/", index=False)
```
